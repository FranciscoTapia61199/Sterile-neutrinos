## Parameters and mass equations used

The parameters and fixed values are listed here separated in NH and IH parameters, same situation for the mass equations (equation 16 at the technical note), for the NH all the mass equations are in terms of $m_1$ and for the IH, in terms of $m_3$, here are not only listed the best fit values, there are also listed the +3/-3 $\sigma$ values for all the masses to get not only the best fit value plot, but also including the uncertainties for the plot.

Note that we are including $s^2_{15}$ and $\Delta m^2_{51}$ as $\Delta m^2_{\text{New}}$ to add $m_{light}$ dependence to $m_{5}$

```python
#Normal hierarchy parameters (Best Fit)

NHs12=0.307
NHs13=0.022
NHc12=0.693
NHc13=0.978
NHs14=0.023
NHc14=0.977
NHs15=0.02
NHc15=0.98

solarmass=0.0000753
atmmass=0.0025283
LSNDmass=1.78
Newmass=0.87

#solarmass=del_m21^2
#atmmass=del_m31^2

#Normal hierarchy parameters (3 sigma)

maxNHs12=0.346
minNHs12=0.268

maxNHs13=0.0241
minNHs13=0.0199

maxNHc12=0.732
minNHc12=0.654

maxNHc13=0.9801
minNHc13=0.9759

maxNHs14=0.04
minNHs14=0.006

maxNHc14=0.96
minNHc14=0.994

maxNHs15=0.035
minNHs15=0.005

maxNHc15=0.965
minNHc15=0.995

maxsolarmass=0.0000807
minsolarmass=0.0000699

maxatmmass=0.0025631
minatmmass=0.0024935

maxLSNDmass=2.01
minLSNDmass=1.61

maxNewmass=0.97
minNewmass=0.77

NHm5=numpy.linspace(0.0001,0.001,2000) 
NHm5=numpy.append(NHm5, numpy.linspace(0.001,0.01,2000))
NHm5=numpy.append(NHm5, numpy.linspace(0.01,0.1,2000))
NHm5=numpy.append(NHm5, numpy.linspace(0.1,1,2000))
NHm5=numpy.append(NHm5, numpy.linspace(1,10,2000))

maxNHm1=(NHm5**2+maxNewmass)**(1/2)
NHm1=(NHm5**2+Newmass)**(1/2)
minNHm1=(NHm5**2+minNewmass)**(1/2)

maxNHm2=(NHm5**2+maxNewmass+maxsolarmass)**(1/2)
NHm2=(NHm5**2+Newmass+solarmass)**(1/2)
minNHm2=(NHm5**2+minNewmass+minsolarmass)**(1/2)

maxNHm3=(NHm5**2+maxNewmass+maxatmmass)**(1/2)
NHm3=(NHm5**2+Newmass+atmmass)**(1/2)
minNHm3=(NHm5**2+minNewmass+minatmmass)**(1/2)

maxNHm4=(NHm5**2+maxLSNDmass+maxNewmass)**(1/2)
NHm4=(NHm5**2+LSNDmass+Newmass)**(1/2)
minNHm4=(NHm5**2+minLSNDmass+minNewmass)**(1/2)
```

```python
#Inverted hierarchy parameters (Best Fit)

IHs12=0.307
IHs13=0.022
IHc12=0.693
IHc13=0.978
IHs14=0.023
IHc14=0.977
IHs15=0.02
IHc15=0.98

Isolarmass=0.0000753
Iatmmass=0.0025283
ILSNDmass=1.78
INewmass=0.87

#solarmass=del_m21^2
#atmmass=del_m31^2

#Inverted hierarchy parameters (3 sigma)

maxIHs12=0.346
minIHs12=0.268

maxIHs13=0.0241
minIHs13=0.0199

maxIHc12=0.732
minIHc12=0.654

maxIHc13=0.9801
minIHc13=0.9759

maxIHs14=0.04
minIHs14=0.006

maxIHc14=0.96
minIHc14=0.994

maxIHs15=0.035
minIHs15=0.005

maxIHc15=0.965
minIHc15=0.995

maxIsolarmass=0.0000807
minIsolarmass=0.0000699

maxIatmmass=0.0025631
minIatmmass=0.0024935

maxILSNDmass=2.01
minILSNDmass=1.61

maxINewmass=0.97
minINewmass=0.77

IHm5=numpy.linspace(0.0001,0.001,2000) 
IHm5=numpy.append(IHm5, numpy.linspace(0.001,0.01,2000))
IHm5=numpy.append(IHm5, numpy.linspace(0.01,0.1,2000))
IHm5=numpy.append(IHm5, numpy.linspace(0.1,1,2000))
IHm5=numpy.append(IHm5, numpy.linspace(1,10,2000))

maxIHm1=(IHm5**2+maxINewmass)**(1/2)
IHm1=(NHm5**2+INewmass)**(1/2)
minIHm1=(NHm5**2+minINewmass)**(1/2)

maxIHm2=(IHm5**2+maxINewmass+maxIsolarmass)**(1/2)
IHm2=(IHm5**2+INewmass+Isolarmass)**(1/2)
minIHm2=(IHm5**2+minINewmass+minIsolarmass)**(1/2)

maxIHm3=(IHm5**2+maxINewmass-maxIatmmass)**(1/2)
IHm3=(IHm5**2+INewmass-Iatmmass)**(1/2)
minIHm3=(IHm5**2+minINewmass-minIatmmass)**(1/2)

maxIHm4=(IHm5**2+maxILSNDmass+maxINewmass)**(1/2)
IHm4=(IHm5**2+ILSNDmass+INewmass)**(1/2)
minIHm4=(IHm5**2+minILSNDmass+minINewmass)**(1/2)
```

## Plot generator

This is the most important part of the code, on this cell is located the original $\langle{m_{ee}}\rangle$ formula and is written in terms of $m_{light}$ that is $m_{1}$ for NH and $m_{3}$ for IH, since there are infinite many possible combinations of parameters $\phi_{12}$, $\phi_{13}$, $\phi_{14}$, $\phi_{15}$ (equation 11 of the technical note), instead of running a code with infinite many curves, the code will just plug the curves that will bound all the possible angle combinations and fill the space between each curve to optimize compilation time and computer resources, the angles that created the boundaries were calculated previously using an optimization method (details and other equations used are located at the technical note). 

Note, these cells also include a lobster plot for 3 neutrinos with its best fit values to make a comparison to the new 1+3+I neutrino plot.

```python
plt.figure(figsize=(9,6))

NHmee11=abs(NHc12*NHc13*NHc14*NHc15*NHm1 + 
              NHs12*NHc13*NHc14*NHc15*NHm2*cmath.exp(complex(0,0)) +
              NHs13*NHc14*NHc15*NHm3*cmath.exp(complex(0,0)) +
              NHs14*NHc15*NHm4*cmath.exp(complex(0,0)) + 
              NHs15*NHm5*cmath.exp(complex(0,0)))

NHmee12=abs(NHc12*NHc13*NHc14*NHc15*NHm1 + 
              NHs12*NHc13*NHc14*NHc15*NHm2*cmath.exp(complex(0,numpy.pi)) +
              NHs13*NHc14*NHc15*NHm3*cmath.exp(complex(0,numpy.pi)) +
              NHs14*NHc15*NHm4*cmath.exp(complex(0,numpy.pi)) + 
              NHs15*NHm5*cmath.exp(complex(0,numpy.pi)))


maxNHmee11=abs(minNHc12*minNHc13*minNHc14*minNHc15*maxNHm1 + 
              maxNHs12*minNHc13*minNHc14*minNHc15*maxNHm2*cmath.exp(complex(0,0)) +
              maxNHs13*minNHc14*minNHc15*maxNHm3*cmath.exp(complex(0,0)) +
              maxNHs14*minNHc15*maxNHm4*cmath.exp(complex(0,0)) + 
              maxNHs15*NHm5*cmath.exp(complex(0,0)))

maxNHmee12=abs(minNHc12*minNHc13*minNHc14*minNHc15*maxNHm1 + 
              maxNHs12*minNHc13*minNHc14*minNHc15*maxNHm2*cmath.exp(complex(0,numpy.pi)) +
              maxNHs13*minNHc14*minNHc15*maxNHm3*cmath.exp(complex(0,numpy.pi)) +
              maxNHs14*minNHc15*maxNHm4*cmath.exp(complex(0,numpy.pi)) + 
              maxNHs15*NHm5*cmath.exp(complex(0,numpy.pi)))
               

minNHmee11=abs(maxNHc12*maxNHc13*maxNHc14*maxNHc15*minNHm1 + 
              minNHs12*maxNHc13*maxNHc14*maxNHc15*minNHm2*cmath.exp(complex(0,0)) +
              minNHs13*maxNHc14*maxNHc15*minNHm3*cmath.exp(complex(0,0)) +
              minNHs14*maxNHc15*minNHm4*cmath.exp(complex(0,0)) + 
              minNHs15*NHm5*cmath.exp(complex(0,0)))

minNHmee12=abs(maxNHc12*maxNHc13*maxNHc14*maxNHc15*minNHm1 + 
              minNHs12*maxNHc13*maxNHc14*maxNHc15*minNHm2*cmath.exp(complex(0,numpy.pi)) +
              minNHs13*maxNHc14*maxNHc15*minNHm3*cmath.exp(complex(0,numpy.pi)) +
              minNHs14*maxNHc15*minNHm4*cmath.exp(complex(0,numpy.pi)) + 
              minNHs15*NHm5*cmath.exp(complex(0,numpy.pi)))


#plt.plot(NHm5,NHmee11,color="red",linestyle="-",linewidth=2, alpha=1,zorder=16)
#plt.plot(NHm5,NHmee12,color="gray",linestyle="-",linewidth=2, alpha=1,zorder=16)

#plt.plot(NHm5,maxNHmee11,color="red",linestyle="--",linewidth=2, alpha=1,zorder=16)
#plt.plot(NHm5,maxNHmee12,color="gray",linestyle="--",linewidth=2, alpha=1,zorder=16)

#plt.plot(NHm5,minNHmee11,color="red",linestyle="-.",linewidth=2, alpha=1,zorder=16)
#plt.plot(NHm5,minNHmee12,color="gray",linestyle="-.",linewidth=2, alpha=1,zorder=16)

plt.fill_between(NHm5, NHmee11, NHmee12, color="blue",label="(1+3+I) Best fit NH",alpha=1,zorder=14)

plt.fill_between(NHm5, maxNHmee11, maxNHmee12, color="navy",label="(1+3+I) 3$\sigma$ NH",alpha=1,zorder=13)
               
plt.xscale("log")
plt.yscale("log")

plt.xlim(0.0001,10)
plt.ylim(0.0001,10)
plt.tight_layout()
plt.title("Normal ordering", fontsize=18)
plt.xlabel("$m_{light}$ in eV",fontsize=16)
plt.ylabel("$m_{ee}$ in eV",fontsize=16)
plt.legend(fontsize=10,loc=2)
plt.show()
```
